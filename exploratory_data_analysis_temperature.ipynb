```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_temp = pd.read_csv('mean_temp.csv', index_col=0)
```

```python
df_temp.index = pd.to_datetime(df_temp.index)
```

```python
df_temp = df_temp.replace(to_replace= 99.9, value= pd.np.nan).dropna(axis = 1, how = 'any')
```

```python
df_temp.shape
```

```python
sns.set()
```

```python
cols = df_temp.columns
```

```python
df_temp_resampled = df_temp.resample(rule= 'M')
```

```python
df_temp_monthy_avg = df_temp_resampled.mean()
```

```python
df_temp_monthy_avg.head()
```

```python
_, ax = plt.subplots(1, 1, figsize = (12, 9))
df_temp_monthy_avg[cols[2]].plot.line()
```
